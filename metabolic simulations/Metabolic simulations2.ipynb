## Stability analysis

#### Puntos de equilibrio

```python
import numpy as np
import pandas as pd
from scipy.optimize import fsolve
from numpy.linalg import eigvals

# -------------------------
# PARÁMETROS DEL MODELO
# -------------------------
d_1, d_2 = 0.008, 0.005
c4 = 1.88
c5 = 0.684555
c1 = 0.05425 
c2 = 0.0515
c3 = 1.99
c6 = 5.06

# -------------------------
# SISTEMA EN EQUILIBRIO
# -------------------------
def sistema_equilibrio(y):
    x1, x2, O2, CO2 = y
    dx1 = c1 * CO2  * x1 * (1 - x1 - x2) - d_1 * x1
    dx2 = c2 * O2  * x2 * (1 - x1 - x2) - d_2 * x2
    dO2 = c3 * x1 - c4 * O2* x2 * (1 - x1 - x2) - 0.1 * O2
    dCO2 = -c5 * CO2 * x1 * (1 - x1 - x2) + c6 * x2 - 0.6 * CO2
    return [dx1, dx2, dO2, dCO2]

# -------------------------
# JACOBIANO DERIVADO A MANO
# -------------------------

def jacobiano_manual(y):
    x1, x2, O2, CO2 = y

    J = np.zeros((4, 4))

    # Parciales de dx1/dt
    J[0, 0] = c1 * CO2 * (1 - 2 * x1 - x2) - d_1
    J[0, 1] = -c1 * CO2 * x1
    J[0, 2] = 0
    J[0, 3] = c1 * x1 * (1 - x1 - x2)

    # Parciales de dx2/dt
    J[1, 0] = -c2 * O2 * x2
    J[1, 1] = c2 * O2 * (1 - 2 * x2 - x1) - d_2
    J[1, 2] = c2 * x2 * (1 - x1 - x2)
    J[1, 3] = 0

    # Parciales de dO2/dt
    J[2, 0] = c3 + c4 * O2 * x2
    J[2, 1] = c4 * O2 * (x1 + 2 * x2 - 1)
    J[2, 2] = c4 * x2 * (x1 + x2 - 1) - do2
    J[2, 3] = 0

    # Parciales de dCO2/dt
    J[3, 0] = CO2 * c5 * (2 * x1 + x2 - 1)
    J[3, 1] = CO2 * c5 * x1 + c6
    J[3, 2] = 0
    J[3, 3] = c5 * x1 * (x1 + x2 - 1) - dco2

    return J

# -------------------------
# CONDICIONES INICIALES
# -------------------------
initial_conditions = [
    [0.3, 0.2, 1.0, 1.0],
    [0.7, 0.05, 2.0, 1.5],
    [0.05, 0.7, 1.5, 2.0],
    [0.1, 0.1, 0.5, 0.5],
    [0.4, 0.4, 3.0, 3.0],
    [0.6, 0.2, 1.2, 0.8],
    [0.2, 0.6, 0.8, 1.2],
    [0.005650, 0.0021, 0.10846, 0.017],
    [0.7871, 0.0431, 15.53,  0.2542],
    [0.7996, 0.0543, 15.77, 0.3540]
]

# -------------------------
# BÚSQUEDA REFINADA
# -------------------------
found_equilibria = []

for y0 in initial_conditions:
    try:
        sol, info, ier, msg = fsolve(sistema_equilibrio, y0, xtol=1e-12, maxfev=10000, full_output=True)
        residuo = np.linalg.norm(sistema_equilibrio(sol))
        rounded = tuple(np.round(sol, 5))
        if ier == 1 and residuo < 1e-8:
            if not any(np.allclose(rounded, eq, atol=1e-12) for eq in found_equilibria):
                found_equilibria.append(rounded)
    except:
        continue

# Añadir el equilibrio trivial si no está
trivial = (0.0, 0.0, 0.0, 0.0)
if not any(np.allclose(trivial, eq, atol=1e-12) for eq in found_equilibria):
    found_equilibria.append(trivial)

# -------------------------
# ESTABILIDAD DE CADA PUNTO
# -------------------------
estabilidades = []
for eq in found_equilibria:
    J = jacobiano_manual(eq)
    autovalores = eigvals(J)
    if np.all(np.real(autovalores) < 0):
        estabilidades.append("Estable")
    elif np.any(np.real(autovalores) > 0):
        estabilidades.append("Inestable")
    else:
        estabilidades.append("Saddle")

# -------------------------
# RESULTADOS
# -------------------------
df_result = pd.DataFrame(found_equilibria, columns=["x1", "x2", "O2", "CO2"])
df_result["Estabilidad"] = estabilidades
print("Puntos de equilibrio y su estabilidad:")
print(df_result)
```

```python
# -------------------------
# MOSTRAR AUTORES DE CADA EQUILIBRIO
# -------------------------
print("\nAutovalores de cada equilibrio:\n")
for i, eq in enumerate(found_equilibria):
    J = jacobiano_manual(eq)
    autovalores = eigvals(J)
    print(f"Equilibrio {i+1}: {np.round(eq, 5)}")
    print(f"  Estabilidad: {estabilidades[i]}")
    print(f"  Autovalores: {np.round(autovalores, 5)}\n")
```

## Stability analysis based on O2 and CO2 extraction rates 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# -------------------------
# PARÁMETROS DEL MODELO
# -------------------------
# Condiciones iniciales
y0 = [0.3, 0.3, 1.0, 1.0]  # x1, x2, O2, CO2
t_span_long = (0, 5000)
t_eval_long = np.linspace(*t_span_long, 10000)

# Parámetros del sistema
d_1, d_2 = 0.008, 0.005
c1, c2, c3, c4 = 0.05425, 0.0515, 1.99, 1.88
c5, c6 = 0.684555, 5.06
dco2 = 0.6  # extracción de CO2 fija

# -------------------------
# SISTEMA DINÁMICO MODIFICADO
# -------------------------
def dynamic_system_param(t, y, do2, dco2):
    x1, x2, O2, CO2 = y
    dx1 = c1 * (CO2 / (1 + CO2)) * x1 * (1 - x1 - x2) - d_1 * x1
    dx2 = c2 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - d_2 * x2
    dO2 = c3 * x1 - c4 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - do2 * O2
    dCO2 = -c5 * (CO2 / (1 + CO2)) * x1 * (1 - x1 - x2) + c6 * x2 - dco2 * CO2
    return [dx1, dx2, dO2, dCO2]

# -------------------------
# ANÁLISIS EXTENDIDO DE TIPPING POINT
# -------------------------
do2_vals_ext = np.linspace(0, 4.5, 100)
x1_final_long = []
x2_final_long = []

for val in do2_vals_ext:
    sol = solve_ivp(dynamic_system_param, t_span_long, y0, args=(val, dco2), t_eval=[t_eval_long[-1]], method='BDF')
    x1_final_long.append(sol.y[0, -1])
    x2_final_long.append(sol.y[1, -1])

# -------------------------
# DETECCIÓN DEL TIPPING POINT
# -------------------------
tipping_index = None
for i in range(len(do2_vals_ext)):
    if x1_final_long[i] < 1e-3 or x2_final_long[i] < 1e-3:
        tipping_index = i
        break
tipping_point = do2_vals_ext[tipping_index] if tipping_index is not None else None

# -------------------------
# GRÁFICO EMBELLECIDO
# -------------------------
plt.figure(figsize=(10, 6))

# Colorear zonas
plt.axvspan(do2_vals_ext[0], tipping_point, color='lightgreen', alpha=0.3, label='Zona viable')
plt.axvspan(tipping_point, do2_vals_ext[-1], color='lightcoral', alpha=0.3, label='Zona de colapso')

# Curvas de biomasa
plt.plot(do2_vals_ext, x1_final_long, label='x₁ final (Chlorella)', color='green', linewidth=2)
plt.plot(do2_vals_ext, x2_final_long, label='x₂ final (Bacillus)', color='blue', linewidth=2)

# Línea de tipping point
plt.axvline(tipping_point, color='black', linestyle='--', linewidth=2, label=f'Tipping point ≈ {tipping_point:.2f}')

# Estética general
plt.axhline(0, color='gray', linestyle=':')
plt.xlabel('Tasa de extracción de O₂ (do₂)', fontsize=12)
plt.ylabel('Concentración final de biomasa', fontsize=12)
plt.title('Extinción en función de la extracción de O₂ (t = 5000)', fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# -------------------------
# PARÁMETROS Y CONDICIONES
# -------------------------
y0 = [0.3, 0.3, 1.0, 1.0]  # x1, x2, O2, CO2
t_sim = 5000
d_1, d_2 = 0.008, 0.005
c1, c2, c3, c4 = 0.05425, 0.0515, 1.99, 1.88
c5, c6 = 0.684555, 5.06
dco2_fixed = 0.6

# -------------------------
# SISTEMA DINÁMICO MODIFICADO
# -------------------------
def dynamic_system_param(t, y, do2, dco2):
    x1, x2, O2, CO2 = y
    dx1 = c1 * (CO2 / (1 + CO2)) * x1 * (1 - x1 - x2) - d_1 * x1
    dx2 = c2 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - d_2 * x2
    dO2 = c3 * x1 - c4 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - do2 * O2
    dCO2 = -c5 * (CO2 / (1 + CO2)) * x1 * (1 - x1 - x2) + c6 * x2 - dco2 * CO2
    return [dx1, dx2, dO2, dCO2]

# -------------------------
# BARRIDO DE VALORES DE do2
# -------------------------
do2_vals = np.linspace(0, 4.5, 100)
x1_final, x2_final, o2_final, co2_final = [], [], [], []

for do2 in do2_vals:
    sol = solve_ivp(dynamic_system_param, (0, t_sim), y0, args=(do2, dco2_fixed), t_eval=[t_sim], method='BDF')
    x1_final.append(sol.y[0, -1])
    x2_final.append(sol.y[1, -1])
    o2_final.append(sol.y[2, -1])
    co2_final.append(sol.y[3, -1])

# -------------------------
# DETECCIÓN DEL TIPPING POINT
# -------------------------
threshold = 1e-3
tipping_index = next((i for i, (x1, x2) in enumerate(zip(x1_final, x2_final)) if x1 < threshold or x2 < threshold), None)
tipping_point = do2_vals[tipping_index] if tipping_index is not None else None

# -------------------------
# GRÁFICA 1: Biomasa vs do2
# -------------------------
plt.figure(figsize=(10, 6))
plt.axvspan(do2_vals[0], tipping_point, color='lightgreen', alpha=0.3, label='Zona viable')
plt.axvspan(tipping_point, do2_vals[-1], color='lightcoral', alpha=0.3, label='Zona de colapso')
plt.plot(do2_vals, x1_final, label='x₁ final (Chlorella)', color='green', linewidth=2)
plt.plot(do2_vals, x2_final, label='x₂ final (Bacillus)', color='blue', linewidth=2)
plt.axvline(tipping_point, color='black', linestyle='--', linewidth=2, label=f'Tipping point ≈ {tipping_point:.2f}')
plt.axhline(0, color='gray', linestyle=':')
plt.xlabel('Tasa de extracción de O₂ Eo₂)', fontsize=12)
plt.ylabel('Concentración final de biomasa', fontsize=12)
plt.title('Biomasa final vs extracción de O₂ (Eco₂ fija)', fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('tipping_point.png', dpi=300, bbox_inches='tight')
plt.show()

# -------------------------
# GRÁFICA 2: Gases vs do2
# -------------------------
plt.figure(figsize=(10, 6))
plt.axvspan(do2_vals[0], tipping_point, color='lightgreen', alpha=0.3, label='Zona viable')
plt.axvspan(tipping_point, do2_vals[-1], color='lightcoral', alpha=0.3, label='Zona de colapso')
plt.plot(do2_vals, o2_final, label='O₂ final', color='orange', linewidth=2)
plt.plot(do2_vals, co2_final, label='CO₂ final', color='brown', linewidth=2)
plt.axvline(tipping_point, color='black', linestyle='--', linewidth=2, label=f'Tipping point ≈ {tipping_point:.2f}')
plt.axhline(0, color='gray', linestyle=':')
plt.xlabel('Tasa de extracción de O₂ (Eo₂)', fontsize=12)
plt.ylabel('Concentración final de gases', fontsize=12)
plt.title('Concentraciones de gases vs extracción de O₂ (Eco₂ fija)', fontsize=14)
plt.ylim(0, 8)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('tipping_point1.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Defining a lineal EDOs system (instead of Monod) and comparing its performance regarding O2 extraction rate to our first mathematical approximation

## Evaluating system resistance: attraction slopes of equilibrium points: coexistance or extintion

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros
d_1, d_2 = 0.008, 0.005
c4 = 1.88
c5 = 0.684555
c1 = 0.05425
c2 = 0.0515
c3 = 1.99
c6 = 5.06

# Sistema de ecuaciones
def hiperciclo(t, y):
    x1, x2, O2, CO2 = y
    dx1 = c1 * (CO2 / (0.1 + CO2)) * x1 * (1 - x1 - x2) - x1 * d_1
    dx2 = c2 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - x2 * d_2
    dO2 = c3 * x1 - c4 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - 2 * O2
    dCO2 = -c5 * (CO2 / (0.1 + CO2)) * x1 * (1 - x1 - x2) + c6 * x2 - 15 * CO2
    return [dx1, dx2, dO2, dCO2]
```

```python
# Rango de condiciones iniciales para x1 y x2
x_vals = np.linspace(0, 0.5, 1000)
y_vals = np.linspace(0, 0.5, 1000)

# Matriz para guardar si sobrevive o no (0: muere, 1: sobrevive)
supervivencia = np.zeros((len(x_vals), len(y_vals)))
biomasa_x1 = np.zeros_like(supervivencia)
biomasa_x2 = np.zeros_like(supervivencia)

# Condiciones iniciales fijas de gases
O2_0 = 1.0
CO2_0 = 1.0

# Tiempo de simulación
t_span = (0, 5000)
t_eval = np.linspace(*t_span, 1000)

# Simulaciones
for i, x1_0 in enumerate(x_vals):
    for j, x2_0 in enumerate(y_vals):
        sol = solve_ivp(hiperciclo, t_span, [x1_0, x2_0, O2_0, CO2_0], t_eval=t_eval, method='BDF')
        x1_final, x2_final = sol.y[0, -1], sol.y[1, -1]

        # Consideramos que "muere" si x1 y x2 tienden a cero
        if x1_final < 0.01 and x2_final < 0.01:
            supervivencia[i, j] = 0
        else:
            supervivencia[i, j] = 1
        
        biomasa_x1[i, j] = x1_final
        biomasa_x2[i, j] = x2_final
```

### Mathematical analysis of lineal approximation

## Análisis matemático de la tasa crítica de extracción de oxígeno 

Queremos calcular el valor máximo de \( do_2 \) (la tasa de extracción de oxígeno) que permite que el sistema mantenga la coexistencia estable de *Chlorella* (\( x_1 \)) y *Bacillus* (\( x_2 \)).

---

### Paso 1: Condición de coexistencia

En equilibrio:

$$
\frac{dx_1}{dt} = 0 \quad \text{y} \quad \frac{dx_2}{dt} = 0
$$

De las dos ecuaciones:

$$
c_1 \cdot CO_2 \cdot (1 - x_1 - x_2) = d_1
$$

$$
c_2 \cdot O_2 \cdot (1 - x_1 - x_2) = d_2
$$

Igualando el segundo miembro:

$$
\frac{d_1}{c_1 \cdot CO_2} = \frac{d_2}{c_2 \cdot O_2} \Rightarrow
O_2 = \frac{d_2 \cdot c_1 \cdot CO_2}{d_1 \cdot c_2}
$$

---

### Paso 2: Ecuación del oxígeno en equilibrio

La tercera ecuación del sistema, en equilibrio, es:

$$
\frac{dO_2}{dt} = c_3 x_1 - c_4 \cdot O_2 \cdot x_2 \cdot (1 - x_1 - x_2) - do_2 \cdot O_2 = 0
$$

Sustituimos las expresiones anteriores:
Entonces:

$$
0 = c_3 x_1 - c_4 \cdot \left( \frac{d_2 c_1 CO_2}{d_1 c_2} \right) \cdot x_2 \cdot \left( \frac{d_1}{c_1 CO_2} \right) - do_2 \cdot \left( \frac{d_2 c_1 CO_2}{d_1 c_2} \right)
$$

---

### Paso 3: Simplificamos

$$
0 = c_3 x_1 - \frac{c_4 d_2 x_2}{c_2} - do_2 \cdot \left( \frac{d_2 c_1 CO_2}{d_1 c_2} \right)
$$

Despejamos \( do_2 \):

$$
do_2 = \frac{d_1 c_2}{d_2 c_1 CO_2} \cdot \left( c_3 x_1 - \frac{c_4 d_2 x_2}{c_2} \right)
$$

---

### Paso 4: Sustituimos los valores reales
- \( x_1 = 0.3 \)
- \( x_2 = 0.2 \)
- \( CO_2 = 0.173 \)
- \( d_1 = 0.008 \)
- \( d_2 = 0.005 \)
- \( c_1 = 0.05425 \)
- \( c_2 = 0.0515 \)
- \( c_3 = 1.99 \)
- \( c_4 = 1.88 \)

Cálculos:
$$
( d_1 c_2 = 0.000412 )
$$
$$
( d_2 c_1 CO_2 = 0.00004693 )
$$
$$
( c_3 x_1 = 0.597 )
$$

$$
( \frac{c_4 d_2 x_2}{c_2} \approx 0.0365 )
$$
$$
 Diferencia: ( 0.597 - 0.0365 = 0.5605 )
$$ 

Resultado final:

$$
do_2^{\text{crítico}} = \frac{0.000412}{0.00004693} \cdot 0.5605 \approx 8.78 \cdot 0.5605 \approx \boxed{4.92}
$$

---

### Conclusión

- Si \( do_2 < 4.92 \): el sistema puede mantener la coexistencia estable.
- Si \( do_2 > 4.92 \): se extrae demasiado oxígeno y el sistema colapsa.
- En \( do_2 = 4.92 \): ocurre una bifurcación crítica que marca el límite de estabilidad.



##  Desarrollo matemático: Puntos de equilibrio del sistema

Queremos encontrar las expresiones simbólicas de los puntos de equilibrio:

- $x_1$: biomasa de *Chlorella*
- $x_2$: biomasa de *Bacillus*
- $O_2$: concentración de oxígeno
- $CO_2$: concentración de dióxido de carbono

---

###  Ecuaciones en equilibrio

Sistema:

$$
\begin{cases}
c_1 \cdot CO_2 \cdot x_1 \cdot (1 - x_1 - x_2) - d_1 \cdot x_1 = 0 \\
c_2 \cdot O_2 \cdot x_2 \cdot (1 - x_1 - x_2) - d_2 \cdot x_2 = 0 \\
c_3 \cdot x_1 - c_4 \cdot O_2 \cdot x_2 \cdot (1 - x_1 - x_2) - do_2 \cdot O_2 = 0 \\
-c_5 \cdot CO_2 \cdot x_1 \cdot (1 - x_1 - x_2) + c_6 \cdot x_2 - d_{CO_2} \cdot CO_2 = 0
\end{cases}
$$

---

### Paso 1: Igualamos las dos expresiones de $1 - x_1 - x_2$

De las dos primeras ecuaciones, despejamos:

$$
1 - x_1 - x_2 = \frac{d_1}{c_1 CO_2} = \frac{d_2}{c_2 O_2}
\quad \Rightarrow \quad
O_2 = \frac{d_2 \cdot c_1 \cdot CO_2}{d_1 \cdot c_2}
$$

---

### Paso 2: Sustituimos $O_2$ y $1 - x_1 - x_2$

Reemplazamos:

- $O_2 = \dfrac{d_2 \cdot c_1 \cdot CO_2}{d_1 \cdot c_2}$
- $1 - x_1 - x_2 = \dfrac{d_1}{c_1 \cdot CO_2}$

en las ecuaciones 3 y 4, y resolvemos para $x_1$ y $x_2$:

---

###  Soluciones de equilibrio

$$
x_1 = \frac{
CO_2 \cdot c_1 \left( c_1 c_6 d_2 \cdot do_2 + c_4 d_1 d_2 \cdot d_{CO_2} \right)
}{
c_1 c_2 c_3 c_6 d_1 - c_4 c_5 d_1^2 d_2
}
$$

$$
x_2 = \frac{
CO_2 \cdot c_1 \left( c_2 c_3 \cdot d_{CO_2} + c_5 d_2 \cdot do_2 \right)
}{
c_1 c_2 c_3 c_6 - c_4 c_5 d_1 d_2
}
$$

### Bifurcation diagram: oxygen extraction + x1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from collections import defaultdict

# Parámetros del sistema
d_1, d_2 = 0.008, 0.005
c1, c2 = 0.05425, 0.0515
c3, c4 = 1.99, 1.88
c5, c6 = 0.684555, 5.06
dco2 = 0.6

# Sistema en equilibrio
def sistema_equilibrio(y, do2):
    x1, x2, O2, CO2 = y
    eq1 = c1 * CO2 * x1 * (1 - x1 - x2) - d_1 * x1
    eq2 = c2 * O2 * x2 * (1 - x1 - x2) - d_2 * x2
    eq3 = c3 * x1 - c4 * O2 * x2 * (1 - x1 - x2) - do2 * O2
    eq4 = -c5 * CO2 * x1 * (1 - x1 - x2) + c6 * x2 - dco2 * CO2
    return [eq1, eq2, eq3, eq4]

# Jacobiano del sistema
def jacobiano_manual(y, do2):
    x1, x2, O2, CO2 = y
    J = np.zeros((4, 4))
    J[0, 0] = c1 * CO2 * (1 - 2 * x1 - x2) - d_1
    J[0, 1] = -c1 * CO2 * x1
    J[0, 2] = 0
    J[0, 3] = c1 * x1 * (1 - x1 - x2)
    J[1, 0] = -c2 * O2 * x2
    J[1, 1] = c2 * O2 * (1 - 2 * x2 - x1) - d_2
    J[1, 2] = c2 * x2 * (1 - x1 - x2)
    J[1, 3] = 0
    J[2, 0] = c3 + c4 * O2 * x2
    J[2, 1] = c4 * O2 * (x1 + 2 * x2 - 1)
    J[2, 2] = c4 * x2 * (x1 + x2 - 1) - do2
    J[2, 3] = 0
    J[3, 0] = CO2 * c5 * (2 * x1 + x2 - 1)
    J[3, 1] = CO2 * c5 * x1 + c6
    J[3, 2] = 0
    J[3, 3] = c5 * x1 * (x1 + x2 - 1) - dco2
    return J

# Valores del parámetro de bifurcación
do2_vals = np.linspace(0.001, 5, 250)

# Condiciones iniciales para buscar soluciones estables
condiciones_estables = [
    [0.3, 0.2, 1.0, 1.0],
    [0.25, 0.25, 1.2, 1.2],
    [0.4, 0.3, 1.5, 2.0],
    [0.2, 0.1, 1.0, 0.8],
    [0.15, 0.1, 0.9, 0.9],
]

# Condiciones iniciales para buscar soluciones inestables
condiciones_inestables = [
    [0.05, 0.05, 0.3, 0.3],
    [0.1, 0.05, 0.5, 0.5],
    [0.08, 0.12, 0.2, 0.3],
    [0.12, 0.1, 0.6, 0.7],
]

# Inicializamos listas
do2_vals_estable = []
x1_vals_estable = []
do2_vals_inestable = []
x1_vals_inestable = []

# Rama estable
for do2 in do2_vals:
    for y0 in condiciones_estables:
        try:
            sol = fsolve(sistema_equilibrio, y0, args=(do2,))
            sol = np.array(sol, dtype=np.float64)
            x1, x2, O2, CO2 = sol

            if x1 < 1e-3 or x2 < 1e-3 or np.any(sol < 0):
                continue

            J = jacobiano_manual(sol, do2)
            eigvals = np.linalg.eigvals(J)

            if np.all(np.real(eigvals) < 0):
                do2_vals_estable.append(do2)
                x1_vals_estable.append(x1)
                break
        except:
            continue

# Rama inestable
for do2 in do2_vals:
    for y0 in condiciones_inestables:
        try:
            sol = fsolve(sistema_equilibrio, y0, args=(do2,))
            sol = np.array(sol, dtype=np.float64)
            x1, x2, O2, CO2 = sol

            if x1 < 1e-3 or x2 < 1e-3 or np.any(sol < 0):
                continue

            J = jacobiano_manual(sol, do2)
            eigvals = np.linalg.eigvals(J)

            if np.any(np.real(eigvals) > 0):
                do2_vals_inestable.append(do2)
                x1_vals_inestable.append(x1)
                break
        except:
            continue

# Gráfico final
plt.figure(figsize=(10, 5))
plt.title("Diagrama de bifurcación para $x_1$ (Chlorella)")
plt.plot(do2_vals_estable, x1_vals_estable, 'b-', label='Estable (coexistencia)')
plt.plot(do2_vals_inestable, x1_vals_inestable, 'b--', label='Inestable (coexistencia)')
plt.xlabel("$do_2$ (extracción de oxígeno)")
plt.ylabel("$x_1$ (Chlorella en equilibrio)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Bifurcation diagram: oxygen extraction + x2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Parámetros del sistema
d_1, d_2 = 0.008, 0.005
c1, c2 = 0.05425, 0.0515
c3, c4 = 1.99, 1.88
c5, c6 = 0.684555, 5.06
dco2 = 0.6

# Sistema en equilibrio
def sistema_equilibrio(y, do2):
    x1, x2, O2, CO2 = y
    eq1 = c1 * CO2 * x1 * (1 - x1 - x2) - d_1 * x1
    eq2 = c2 * O2 * x2 * (1 - x1 - x2) - d_2 * x2
    eq3 = c3 * x1 - c4 * O2 * x2 * (1 - x1 - x2) - do2 * O2
    eq4 = -c5 * CO2 * x1 * (1 - x1 - x2) + c6 * x2 - dco2 * CO2
    return [eq1, eq2, eq3, eq4]

# Jacobiano del sistema
def jacobiano_manual(y, do2):
    x1, x2, O2, CO2 = y
    J = np.zeros((4, 4))
    J[0, 0] = c1 * CO2 * (1 - 2 * x1 - x2) - d_1
    J[0, 1] = -c1 * CO2 * x1
    J[0, 2] = 0
    J[0, 3] = c1 * x1 * (1 - x1 - x2)
    J[1, 0] = -c2 * O2 * x2
    J[1, 1] = c2 * O2 * (1 - 2 * x2 - x1) - d_2
    J[1, 2] = c2 * x2 * (1 - x1 - x2)
    J[1, 3] = 0
    J[2, 0] = c3 + c4 * O2 * x2
    J[2, 1] = c4 * O2 * (x1 + 2 * x2 - 1)
    J[2, 2] = c4 * x2 * (x1 + x2 - 1) - do2
    J[2, 3] = 0
    J[3, 0] = CO2 * c5 * (2 * x1 + x2 - 1)
    J[3, 1] = CO2 * c5 * x1 + c6
    J[3, 2] = 0
    J[3, 3] = c5 * x1 * (x1 + x2 - 1) - dco2
    return J

# Valores de do2
do2_vals = np.linspace(0.001, 5, 250)

# Condiciones iniciales para encontrar soluciones
condiciones_estables = [
    [0.3, 0.2, 1.0, 1.0],
    [0.25, 0.25, 1.2, 1.2],
    [0.4, 0.3, 1.5, 2.0],
    [0.2, 0.1, 1.0, 0.8],
    [0.15, 0.1, 0.9, 0.9],
]

condiciones_inestables = [
    [0.05, 0.05, 0.3, 0.3],
    [0.1, 0.05, 0.5, 0.5],
    [0.08, 0.12, 0.2, 0.3],
    [0.12, 0.1, 0.6, 0.7],
]

# Inicializamos listas para x1 y x2
do2_est_x1, x1_est = [], []
do2_inest_x1, x1_inest = [], []
do2_est_x2, x2_est = [], []
do2_inest_x2, x2_inest = [], []

# Estables
for do2 in do2_vals:
    for y0 in condiciones_estables:
        try:
            sol = fsolve(sistema_equilibrio, y0, args=(do2,))
            sol = np.array(sol, dtype=np.float64)
            x1, x2, O2, CO2 = sol
            if x1 < 1e-3 or x2 < 1e-3 or np.any(sol < 0): continue
            J = jacobiano_manual(sol, do2)
            if np.all(np.real(np.linalg.eigvals(J)) < 0):
                do2_est_x1.append(do2)
                x1_est.append(x1)
                do2_est_x2.append(do2)
                x2_est.append(x2)
                break
        except: continue

# Inestables
for do2 in do2_vals:
    for y0 in condiciones_inestables:
        try:
            sol = fsolve(sistema_equilibrio, y0, args=(do2,))
            sol = np.array(sol, dtype=np.float64)
            x1, x2, O2, CO2 = sol
            if x1 < 1e-3 or x2 < 1e-3 or np.any(sol < 0): continue
            J = jacobiano_manual(sol, do2)
            if np.any(np.real(np.linalg.eigvals(J)) > 0):
                do2_inest_x1.append(do2)
                x1_inest.append(x1)
                do2_inest_x2.append(do2)
                x2_inest.append(x2)
                break
        except: continue

# 🔁 GRÁFICO FINAL COMBINADO
plt.figure(figsize=(10, 6))
plt.plot(do2_est_x2, x2_est, 'g-', label='$x_2$ Estable (Bacillus)')
plt.plot(do2_inest_x2, x2_inest, 'g--', label='$x_2$ Inestable (Bacillus)')
plt.xlabel("$do_2$ (extracción de oxígeno)")
plt.ylabel("Biomasa en equilibrio")
plt.title("Diagrama de bifurcación combinado para $x_2$")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Comparation between bifurcation diagramas of the three models: dFBA, monod approximation and lineal approximation. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from tqdm import tqdm

# ================================
# PARÁMETROS Y CONDICIONES INICIALES
# ================================
y0 = [0.3, 0.2, 1, 1]  # Biomasa Chlorella, Bacillus, O2, CO2
t_span = (0, 5000)
t_eval = [t_span[1]]
d_1, d_2 = 0.008, 0.005

chlorella_biomass_rxn = 'Biomass_Cvu_auto_DASH_'
bacillus_biomass_rxn = 'BIOMASS_BS_10'
chlorella_o2_exchange = "DM_o2D_LPAREN_u_RPAREN_"
bacillus_o2_exchange = 'EX_o2_e'

# ================================
# RANGO DE do2 A EXPLORAR
# ================================
do2_vals_ext = np.linspace(0, 5, 1000)

# ================================
# 1. Modelo LINEAL
# ================================
c1, c2, c3, c4 = 0.05425, 0.0515, 1.99, 1.88
c5, c6 = 0.684555, 5.06
dco2 = 0.6

x1_final_linear = []
x2_final_linear = []

for do2 in tqdm(do2_vals_ext, desc="Modelo lineal"):

    def dynamic_system_linear(t, y):
        x1, x2, O2, CO2 = y
        dx1 = c1 * CO2 * x1 * (1 - x1 - x2) - d_1 * x1
        dx2 = c2 * O2 * x2 * (1 - x1 - x2) - d_2 * x2
        dO2 = c3 * x1 - c4 * O2 * x2 * (1 - x1 - x2) - do2 * O2
        dCO2 = -c5 * CO2 * x1 * (1 - x1 - x2) + c6 * x2 - dco2 * CO2
        return [dx1, dx2, dO2, dCO2]

    sol = solve_ivp(
        fun=dynamic_system_linear,
        t_span=t_span,
        y0=y0,
        t_eval=t_eval,
        method='BDF',
        rtol=1e-6,
        atol=1e-8
    )

    x1_final_linear.append(sol.y[0, -1])
    x2_final_linear.append(sol.y[1, -1])

# ================================
# 2. Modelo MONOD
# ================================
x1_final_monod = []
x2_final_monod = []

for do2 in tqdm(do2_vals_ext, desc="Modelo Monod"):

    def dynamic_system_monod(t, y):
        x1, x2, O2, CO2 = y
        dx1 = c1 * (CO2 / (1 + CO2)) * x1 * (1 - x1 - x2) - d_1 * x1
        dx2 = c2 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - d_2 * x2
        dO2 = c3 * x1 - c4 * (O2 / (1 + O2)) * x2 * (1 - x1 - x2) - do2 * O2
        dCO2 = -c5 * (CO2 / (1 + CO2)) * x1 * (1 - x1 - x2) + c6 * x2 - dco2 * CO2
        return [dx1, dx2, dO2, dCO2]

    sol = solve_ivp(
        fun=dynamic_system_monod,
        t_span=t_span,
        y0=y0,
        t_eval=t_eval,
        method='BDF',
        rtol=1e-6,
        atol=1e-8
    )

    x1_final_monod.append(sol.y[0, -1])
    x2_final_monod.append(sol.y[1, -1])
# ================================
# GRÁFICO FINAL: SUPERPOSICIÓN
# ================================
plt.figure(figsize=(14, 8))

# Chlorella
plt.plot(do2_vals_ext, x1_final_linear, label='Chlorella (Lineal)', color='green', linestyle='-')
plt.plot(do2_vals_ext, x1_final_monod, label='Chlorella (Monod)', color='red', linestyle='--')


# Bacillus
plt.plot(do2_vals_ext, x2_final_linear, label='Bacillus (Lineal)', color='blue', linestyle='-')
plt.plot(do2_vals_ext, x2_final_monod, label='Bacillus (Monod)', color='purple', linestyle='--')

# Estética
plt.axhline(0, color='gray', linestyle=':')
plt.xlabel('Tasa de extracción de O₂ (Eo₂)', fontsize=14)
plt.ylabel('Concentración final de biomasa (gDW/L)', fontsize=14)
plt.title('Comparativa de Tipping point: Lineal vs Monod', fontsize=16)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("grafico_tipping_point.png", dpi=300)
plt.show()
```
