Parsing modelSEED's compounds database and annotating ids in the model 

```python
import pandas as pd
import cobra
import re
### Mapear metabolitos desde base de datos modelseed

def clean_modelseed_id(met_id):
    """Elimina sufijos de compartimentos en los metabolitos de ModelSEED (ej. '_c0', '_e0', '_p0', '_b')."""
    return re.sub(r'(_[cepib]\d*|_b)$', '', met_id)  # Remueve "_c0", "_e0", "_p0", "_b"

def extract_db_ids(aliases):
    """Extrae identificadores de varias bases de datos desde 'aliases'."""
    if pd.isna(aliases):  # Si el campo est vaco, retornar None
        return {}

    alias_list = aliases.split("|")  # Separar los alias por '|'
    db_mapping = {}

    for alias in alias_list:
        parts = alias.split(":")
        if len(parts) == 2:
            db_name, db_id = parts
            if "KEGG" in db_name:
                db_mapping["kegg.compound"] = db_id.split(";")[0]  # Tomamos solo el primer KEGG ID
            elif "BiGG" in db_name:
                db_mapping["bigg.metabolite"] = db_id.split(";")[0]
            elif "MetaCyc" in db_name:
                db_mapping["metacyc.compound"] = db_id.split(";")[0]
            elif "AraCyc" in db_name:
                db_mapping["aracyc.compound"] = db_id.split(";")[0]
            elif "ChEBI" in db_name:
                db_mapping["chebi.compound"] = db_id.split(";")[0]

    return db_mapping  # Retorna un diccionario con todas las anotaciones encontradas

# Cargar la base de datos de ModelSEED desde el archivo TSV
compounds_db = pd.read_csv("Compounds.tsv", sep="\t")

# Crear un diccionario de conversin ModelSEED  {KEGG, BiGG, MetaCyc, etc.}
modelseed_to_annotations = {}

for index, row in compounds_db.iterrows():
    db_ids = extract_db_ids(row["aliases"])  # Extraer todas las bases de datos
    if db_ids:
        modelseed_to_annotations[row["id"]] = db_ids  # Asociar ModelSEED ID con las anotaciones



# Buscar identificadores y asignarlos en el modelo
for met in model.metabolites:
    modelseed_id = clean_modelseed_id(met.id)  # Eliminar sufijos de compartimento
    annotations = modelseed_to_annotations.get(modelseed_id, {})  # Buscar en el diccionario

    if annotations:
        if not met.annotation:
            met.annotation = {}  # Asegurar que met.annotation es un diccionario
        met.annotation.update(annotations)  # Agregar todas las anotaciones encontradas
        print(f" Asignadas anotaciones a {met.id}: {annotations}")

# Guardar el modelo con las anotaciones completas
cobra.io.write_sbml_model(model, "dechloromonas_model_full_annotations.sbml")

print(" Anotacin de metabolitos con mltiples bases de datos completada y guardada correctamente.")
```

Parsing reaction database and cross-annotating reactions

```python
import pandas as pd
import cobra
import re

#Anotar reacciones
def clean_modelseed_rxn_id(rxn_id):
    """Elimina sufijos de compartimentos en los IDs de reacciones de ModelSEED (ej. '_c0', '_e0', '_p0', '_b')."""
    return re.sub(r'(_[cepib]\d*|_b)$', '', rxn_id)  # Remueve "_c0", "_e0", "_p0", "_b"

def extract_db_ids(aliases, ec_numbers):
    """Extrae identificadores de varias bases de datos desde 'aliases' y EC numbers."""
    db_mapping = {}

    # Extraer informacin de aliases
    if pd.notna(aliases):  # Si el campo no est vaco
        alias_list = aliases.split("|")  # Separar los alias por '|'
        for alias in alias_list:
            parts = alias.split(":")
            if len(parts) == 2:
                db_name, db_id = parts
                if "KEGG" in db_name:
                    db_mapping["kegg.reaction"] = db_id.split(";")[0]  # Tomamos solo el primer KEGG ID
                elif "BiGG" in db_name:
                    db_mapping["bigg.reaction"] = db_id.split(";")[0]
                elif "MetaCyc" in db_name:
                    db_mapping["metacyc.reaction"] = db_id.split(";")[0]
                elif "AraCyc" in db_name:
                    db_mapping["aracyc.reaction"] = db_id.split(";")[0]

    # Extraer nmero EC (si est disponible)
    if pd.notna(ec_numbers) and ec_numbers != "null":
        db_mapping["ec_number"] = ec_numbers.split(";")[0]  # Tomamos solo el primer EC Number

    return db_mapping  # Retorna un diccionario con todas las anotaciones encontradas

#  Cargar la base de datos de ModelSEED desde el archivo TSV
reactions_db = pd.read_csv("reactions.tsv", sep="\t")

# Crear un diccionario de conversin ModelSEED Reaction ID  {KEGG, BiGG, MetaCyc, etc.}
modelseed_to_annotations = {}

for index, row in reactions_db.iterrows():
    db_ids = extract_db_ids(row["aliases"], row["ec_numbers"])  # Extraer todas las bases de datos
    if db_ids:
        modelseed_to_annotations[row["id"]] = db_ids  # Asociar ModelSEED ID con las anotaciones

# Cargar el modelo metablico generado por ModelSEED
model = model_Anno

# Buscar identificadores y asignarlos en el modelo
for rxn in model.reactions:
    modelseed_rxn_id = clean_modelseed_rxn_id(rxn.id)  # Eliminar sufijos de compartimento
    annotations = modelseed_to_annotations.get(modelseed_rxn_id, {})  # Buscar en el diccionario

    if annotations:
        if not rxn.annotation:
            rxn.annotation = {}  # Asegurar que rxn.annotation es un diccionario
        rxn.annotation.update(annotations)  # Agregar todas las anotaciones encontradas
        print(f" Asignadas anotaciones a {rxn.id}: {annotations}")

# Guardar el modelo con las anotaciones completas
cobra.io.write_sbml_model(model, "dechloromonas_model_rxn_full_annotations.sbml")

print("Anotacin de reacciones con mltiples bases de datos completada y guardada correctamente.")

```

Search PATRIC ids in KEGG database

```python
import cobra
import requests
import pandas as pd
import time
### Anotar identificadores de patric con identificadores KEGG
# Extraer los identificadores de genes de PATRIC desde el modelo
patric_ids = [gene.id for gene in model.genes]
print(f" Se encontraron {len(patric_ids)} genes en el modelo.")

# Funcin para buscar un gen de PATRIC en KEGG
def get_kegg_gene(patric_id):
    """Busca un gen de PATRIC en KEGG y devuelve su ID si est disponible."""
    url = f"http://rest.kegg.jp/find/genes/{patric_id}"
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200 and response.text:
            results = response.text.strip().split("\n")
            if results:
                return results[0].split("\t")[0]  # Extraer el primer ID KEGG encontrado
    except requests.exceptions.RequestException as e:
        print(f"Error en la conexin con KEGG para {patric_id}: {e}")
    return None

#  Buscar en KEGG para cada gen de PATRIC
results = []
for patric_id in patric_ids:
    print(f" Buscando en KEGG para {patric_id}...")
    kegg_gene_id = get_kegg_gene(patric_id)
    results.append([patric_id, kegg_gene_id])
    
    time.sleep(2)  # Pausa de 2 segundos para evitar bloqueos en KEGG

# Guardar los resultados en un archivo CSV
df = pd.DataFrame(results, columns=["PATRIC_ID", "KEGG_Gene_ID"])
output_file = "patric_to_kegg.csv"
df.to_csv(output_file, index=False)

print(f" Archivo guardado: {output_file}")
```

### Ontology annotation

Download ontology database from ModelSEED's github repository

```python
import json
import cobra
import requests

#  Definir la URL base de los archivos en GitHub
github_base_url = "https://raw.githubusercontent.com/ModelSEED/ModelSEEDDatabase/refs/heads/master/Ontologies/"
#  Lista de archivos de ontologa a descargar
ontology_files = {
    "KEGG_KO_ontologyDictionary.json": "KBaseOntology.OntologyDictionary.KEGG_KO_ontologyDictionary.json",
    "KEGG_RXN_ontologyDictionary.json": "KBaseOntology.OntologyDictionary.KEGG_RXN_ontologyDictionary.json",
    "ModelSEED_RXN_ontologyDictionary.json": "KBaseOntology.OntologyDictionary.ModelSEED_RXN_ontologyDictionary.json",
    "MetaCyc_RXN_ontologyDictionary.json": "KBaseOntology.OntologyDictionary.MetaCyc_RXN_ontologyDictionary.json"
}

# Descargar y guardar los archivos JSON
def download_json(file_name, github_file_name):
    url = github_base_url + github_file_name
    response = requests.get(url)
    
    if response.status_code == 200:
        with open(file_name, "w", encoding="utf-8") as f:
            f.write(response.text)
        print(f" {file_name} descargado correctamente.")
    else:
        print(f" Error al descargar {file_name}")

for file_name, github_file_name in ontology_files.items():
    download_json(file_name, github_file_name)

```

### Reaction constraints

The model was built over a complete medium so reactions show no constraints. We apply an arbitrary bound to reaction fluxes, in further steps these bounds should be reviewed based on kinetic parameters and labeling experiments. 

```python
# Identificar reacciones sin restricciones
reactions_without_constraints = [rxn for rxn in model.reactions if rxn.lower_bound == -1000 and rxn.upper_bound == 1000]

print(f"Se han identificado {len(reactions_without_constraints)} reacciones sin restricciones.")
```

```python
# Establecer lmites ms realistas para reacciones metablicas
for rxn in model.reactions:
    if rxn.lower_bound == -1000 and rxn.upper_bound == 1000:
        if "EX_" in rxn.id:  # Para reacciones de intercambio
            rxn.lower_bound = -10  # Limitar la entrada de metabolitos
            rxn.upper_bound = 1000  # Permitir salida si es posible
        elif "DM_" in rxn.id:  # Para reacciones de demanda
            rxn.lower_bound = 0
            rxn.upper_bound = 10  # Controlar la produccin
        else:  # Para reacciones metablicas internas
            rxn.lower_bound = -100  # Limitar la reversibilidad
            rxn.upper_bound = 100

print("Restricciones de flujo aplicadas correctamente.")
```

### Energy cycles and fluxes

Once biomass reaction and precursors have been fixed we find the model has incorrect energy cycles. These are cycles in which ATP is produced on its own, not as byproduct of any reaction. These incorrect energy cycles are produced because directionality and reaction limits have not been correctly established. As we have already mentioned to establish accurate reaction limits kinetic and labeling experiments need to be performed. However we can theoretically predict reversibility of reactions based on their compoundd. We can automatically calculate the reaction's Gibbs energy and based on it determine whether it is reversible or not.

```python
import cobra
import requests
import pandas as pd

# Cargar el modelo metablico en formato SBML
modelo = cobra.io.read_sbml_model("modelo_v13.xml")

# Lista para almacenar los resultados
resultados = []

# Recorrer todas las reacciones del modelo
for rxn in modelo.reactions:
    # Obtener el identificador KEGG desde las anotaciones
    kegg_id = rxn.annotation.get("kegg.reaction", None)
    
    if kegg_id:
        try:
            # Consultar la ecuacin de la reaccin en la API de KEGG
            url = f"https://rest.kegg.jp/get/{kegg_id}"
            response = requests.get(url)
            
            if response.status_code == 200:
                data = response.text.split("\n")

                # Buscar la lnea donde aparece la ecuacin de la reaccin
                equation = "No encontrada"
                for line in data:
                    if line.startswith("EQUATION"):
                        equation = line.replace("EQUATION", "").strip()
                        print(equation)
                        break
            else:
                equation = "Error en la consulta"
            
            # Guardar los resultados
            resultados.append({"Reaccin": rxn.id, "KEGG ID": kegg_id, "Ecuacin": equation})

        except Exception as e:
            resultados.append({"Reaccin": rxn.id, "KEGG ID": kegg_id, "Ecuacin": f"Error: {str(e)}"})

# Convertir la lista en un DataFrame
df_resultados = pd.DataFrame(resultados)

# Guardar los resultados en un archivo CSV
df_resultados.to_csv("reacciones_kegg_ecuaciones.csv", index=False)

# Mostrar los resultados
#import ace_tools as tools
#tools.display_dataframe_to_user(name="Reacciones y Frmulas KEGG", dataframe=df_resultados)
```

```python
import cobra
import pandas as pd
from equilibrator_api import ComponentContribution

# Cargar el modelo metablico en SBML
modelo = cobra.io.read_sbml_model("modelo_v13.xml")

# Inicializar el modelo termodinmico de eQuilibrator
cc = ComponentContribution()

# Lista para almacenar los resultados
resultados = []

# Funcin para encontrar la reaccin en el modelo a partir de KEGG ID
def encontrar_reaccion_por_kegg(kegg_id, modelo):
    for reaction in modelo.reactions:
        if "kegg.reaction" in reaction.annotation and reaction.annotation["kegg.reaction"] == kegg_id:
            return reaction
    return None  # No encontrada

# Iterar sobre cada reaccin en el diccionario
for kegg_id, reac in reactions.items():
    try:
        # Calcular G' estndar
        dG0_prime = cc.standard_dg_prime(reac)

        # Extraer solo el valor numrico (sin unidades ni incertidumbre)
        dG0_prime_value = dG0_prime.magnitude

        # Determinar los lmites de flujo en base a G'
        if abs(dG0_prime_value) < 30:
            lb, ub = -1000, 1000  # Reaccin reversible
        elif dG0_prime_value < -30:
            lb, ub = 0, 1000  # Irreversible hacia adelante
        else:
            lb, ub = -1000, 0  # Irreversible hacia atrs

        # Encontrar la reaccin en el modelo metablico usando KEGG ID
        reaction_obj = encontrar_reaccion_por_kegg(kegg_id, modelo)

        # Aplicar restricciones al modelo en COBRApy si la reaccin fue encontrada
        if reaction_obj:
            reaction_obj.lower_bound = lb
            reaction_obj.upper_bound = ub
            print(f" {reaction_obj.id} (KEGG: {kegg_id}) -> G': {dG0_prime_value:.2f} kJ/mol | Bounds: [{lb}, {ub}]")
        else:
            print(f"No se encontr reaccin en el modelo para KEGG ID {kegg_id}")

        # Guardar los resultados
        resultados.append({
            "KEGG ID": kegg_id, 
            "Reaccin en Modelo": reaction_obj.id if reaction_obj else "No encontrada",
            "G' (kJ/mol)": dG0_prime_value, 
            "Lower Bound": lb, 
            "Upper Bound": ub
        })

    except Exception as e:
        print(f" Error al calcular G' para {kegg_id}: {str(e)}")
        resultados.append({
            "KEGG ID": kegg_id, 
            "Reaccin en Modelo": "Error",
            "G' (kJ/mol)": f"Error: {str(e)}", 
            "Lower Bound": "N/A", 
            "Upper Bound": "N/A"
        })
        continue  # Saltar al siguiente ciclo del bucle sin detener el programa

# Convertir los resultados en un DataFrame
df_resultados = pd.DataFrame(resultados)

# Guardar los resultados en un CSV
df_resultados.to_csv("ajuste_termico_reacciones_equilibrator.csv", index=False)

# Guardar el modelo corregido con las restricciones de flujo ajustadas
cobra.io.write_sbml_model(modelo, "modelo_v14.xml")
```

```python
import cobra

def compare_media(old_medium, new_medium, model):
    """
    Compara el medio antiguo con el nuevo y muestra qu compuestos estn presentes
    en el nuevo medio y cules han sido eliminados.
    
    Parmetros:
    - old_medium: Diccionario del medio antiguo.
    - new_medium: Diccionario del medio nuevo.
    - model: Modelo metablico de COBRA.
    
    Retorna:
    - missing_metabolites: Lista de metabolitos eliminados con su nombre.
    - added_metabolites: Lista de metabolitos aadidos con su nombre.
    """
    old_metabolites = set(old_medium.keys())
    new_metabolites = set(new_medium.keys())
    
    missing_metabolites = old_metabolites - new_metabolites
    added_metabolites = new_metabolites - old_metabolites
    
    missing_info = [(rxn, model.reactions.get_by_id(rxn).name) for rxn in missing_metabolites if rxn in model.reactions]
    added_info = [(rxn, model.reactions.get_by_id(rxn).name) for rxn in added_metabolites if rxn in model.reactions]
    
    return missing_info, added_info

# Ejecutar comparacin
missing_metabolites, added_metabolites = compare_media(old_medium, new_medium, model)

# Imprimir resultados
print("Metabolitos eliminados del medio:")
for rxn_id, met_name in missing_metabolites:
    print(f"{rxn_id}: {met_name}")

print("\nMetabolitos aadidos al medio:")
for rxn_id, met_name in added_metabolites:
    print(f"{rxn_id}: {met_name}")
```
